# Group By (Split Apply Combine)

## goals

- find max goals per team
- find average goals scored per team

```python
import pandas as pd
df = pd.read_csv("stats.csv")
df
```

```python
g = df.groupby('team') # group by team
g
```

```python
#printing team by team
for team, team_df in g:
    print(team)
    print(team_df)
```

```python
# choosing the max in each group
g.max()
```

```python
# average of columns
g.mean()
```

```python
g.describe()
```
