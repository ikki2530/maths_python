```python
# amount of recovered
df_recuperados = df[df['Fecha recuperado'].isna() == False]
len(df_recuperados)
```

```python
df["Actual"] = np.nan
df
```

```python
def normalise_row(row):
    x = float("nan")
    if type(row["Fecha de muerte"]) == type(x) and type(row["Fecha recuperado"]) == type(x):
        return "Enfermo"
    elif type(row["Fecha de muerte"]) != type(x):
        return "Muerto"
    else:
        return "Recuperado"
#     if type(row["Fecha de muerte"]) != type(x) and type(row["Fecha recuperado"]) != type(x):
#         print(row)
#         print("--"*20)
    
df['Actual'] = df.apply(lambda row: normalise_row(row), axis=1)
```

```python
df
```

```python
#plot muerto, enfermo, recuperado
plt.figure(figsize=(14,7))
new_actual = df.groupby('Actual').Actual.agg([len])
new_actual.plot(kind='pie', y='len', autopct="%0.1f%%")
```
