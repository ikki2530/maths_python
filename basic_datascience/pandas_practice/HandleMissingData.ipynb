# Handle Missing Data

```python
import pandas as pd
df = pd.read_csv("weather.csv", parse_dates=["day"]) #reading csv
df.set_index('day', inplace=True) #fixing day as the index
df
```

```python
new_df = df.fillna(0) # replace all NaN with 0
new_df
```

```python
#some data
df.describe()
```

```python
# filling different values for different columns
new_df2 = df.fillna({'temperature': 0, 'windspeed': 0, 'event': 'no event'})
new_df2
```

```python
#fill with the previous value 
new_df3 = df.fillna(method="ffill")
new_df3
```

```python
#fill with the next value 
#new_df3 = df.fillna(method="bfill", axis="columns")
new_df3 = df.fillna(method="bfill", limit=1) # fill only 1 cell with the next value
new_df3
```

```python
#interpolate
new_df4 = df.interpolate(method="time")
new_df4
```

```python
new_df5 = df.dropna() # drop all the columns with at least one NaN
new_df5
```

```python
# if thresh=1 keep the values with at least 1 NaN value
new_df5 = df.dropna(thresh=2) # need at least 2 value value to keep the row
new_df5
```

```python
dt = pd.date_range("01-01-2017", "01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
